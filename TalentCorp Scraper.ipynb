```python
import requests
import random
import pandas as pd
from bs4 import BeautifulSoup
from headers import headers_list
from loading_bar import log_progress
```

```python
url = 'https://star.talentcorp.com.my/emp/search/agBmHe'
url
```

```python
def get_company_info(url):
    company = {}
    page = requests.get(url, headers=random.choice(headers_list))
    soup = BeautifulSoup(page.content, 'html.parser')
    results = soup.find(class_='well')
    # company['Company'] = results.find('h1').text.split('\n')[0]
    info = results.find('div', class_='row')
    for dl in info.findAll('dl'):
        company[dl.find('dt').text.strip()] = dl.find('dd').text.strip('\n')
    return company
```

```python
get_company_info(url)
```

```python
def get_companies_from_page(url):
    companies = []
    page = requests.get(url, headers=random.choice(headers_list))
    soup = BeautifulSoup(page.content, 'html.parser')
    logos = soup.find('div', class_='logo-grid')
    for logo in logos.findAll('a', href=True):
        company = get_company_info(logo['href'])
        company['Company'] = logo['title']
        companies.append(company)
    return companies
```

```python
url = 'https://star.talentcorp.com.my/emp/search'
get_companies_from_page(url)
```

```python
base_url = 'https://star.talentcorp.com.my/emp/search?s%5B0%5D=nam&page={}'
start = 1
end = 20
MAX_PAGE = 188
COMPANY_LIST = []

for i in range(start, end+1):
    print('Processing page {}'.format(i))
    url = base_url.format(i)
    companies = get_companies_from_page(url)
    COMPANY_LIST = COMPANY_LIST + companies
```

```python
df = pd.DataFrame.from_dict(COMPANY_LIST)
df.head()
```

```python
df.to_csv('talentcorp.csv', index=False)
```
