```python
import htmldate
from htmldate import find_date

job_dict = []

for job_elem in job_elems:
    title = job_elem.find('div', attrs={'class': 'sx2jih0 _2j8fZ_0 sIMFL_0 _1JtWu_0'})
    company = job_elem.find('span', attrs={'class': 'sx2jih0 zcydq82c _18qlyvc0 _18qlyvcv _18qlyvc1 _18qlyvc8'})
    location = job_elem.find('span', attrs={'class': 'sx2jih0 zcydq82c zcydq8r iwjz4h0'})
    job_url = "https://www.jobstreet.com.my" + job_elem.find('a', attrs={'class': 'DvvsL_0 _1p9OP'})['href']
    date_posted = date_posted = find_date(job_url)
    # Description
    job_url = "https://www.jobstreet.com.my" + job_elem.find('a', attrs={'class': 'DvvsL_0 _1p9OP'})['href']
    page = requests.get(job_url)
    job_soup = BeautifulSoup(page.content, 'html.parser')
    description = 'No description'
    # Had to do this try/except catch because some of the description returns error because
    # they are None. Might be caused by the different div class name for these jobs or they
    # just don't have a job description. Better to double check
    try:
        description = job_soup.find('div', attrs={'class': 'vDEj0_0'}).get_text()
    except:
        pass
    job_dict.append({
        'title': title.text,
        'company': company.text,
        'location': location.text,
        'url': job_url,
        'date_posted': date_posted,
        'description': description})

df_jobs = pd.DataFrame.from_dict(job_dict)
df_jobs
```

### Iterate all job information from all pages from website

```python
# Try to fetch Job Title for the first 10 pages from the website

url = 'https://www.jobstreet.com.my/en/job-search/data-scientist-jobs/'

for page in range(1,10):
    req = requests.get(url + '/')
    soup = BeautifulSoup(req.text, 'html.parser')
    
    title = soup.find_all('div', attrs={'class': 'sx2jih0 _2j8fZ_0 sIMFL_0 _1JtWu_0'})
    
    for i in range(0,30):
        if page>1:
            print(f"{(i)+page*30}" + title[i].text)
        else:
            print(f"{i}" + title[i].text)
```

```python
# Base url
url = 'https://www.jobstreet.com.my/en/job-search/data-scientist-jobs/'

# When experimenting, it is better that we save the request/soup locally so that we don't
# have to keep scraping the website to get the same thing. If we do too much scraping,
# the website might block our IP address and we will have to wait before we can access them
# again
soup_list = []

# Here we defined the page that we want to iterate through in the for loop
for page in range(1,11):
    # So we should modify our url based on the page that we want to get
    req = requests.get(url + '/' + str(page))
    soup = BeautifulSoup(req.text, 'html.parser')
    soup_list.append(soup)

len(soup_list)
```

```python
# Each item in the soup_list contains the info that we need for each page
# Now we can just experiment with the list instead of scraping the website again and again
soup_list[0]
```

```python
job_dict = []

# Iterate through each page
for soup in soup_list:
    # Instead of trying to get all the title directly, let's try separating our soup into
    # each job
    job_elems = soup.find_all('div', class_='sx2jih0 zcydq84g zcydq83g zcydq86g zcydq85g zcydq81w zcydq82o zcydq8cg zcydq8c4')
    for job in job_elems:
        title = job.find('div', attrs={'class': 'sx2jih0 _2j8fZ_0 sIMFL_0 _1JtWu_0'})
        # Now you can just do it like above, getting the rest of the info
        # (company/location/etc) of each job
        job_dict.append({
            'title': title.text
        })

df_jobs = pd.DataFrame.from_dict(job_dict)
df_jobs
```
