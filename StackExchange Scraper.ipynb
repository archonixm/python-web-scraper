```python
import requests

url ='https://datascience.stackexchange.com/questions'
response = requests.get(url, params={'tab':'newest', 'page':1})
```

```python
response.status_code
```

```python
soup = BeautifulSoup(response.text, 'lxml')
soup

soup.prettify
```

```python
results = soup.find_all('div', attrs={'class': 's-post-summary--content'})
len(results)
```

```python
product=results[0]
```
