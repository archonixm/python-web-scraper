```python
questions = []
full_questions = []
for page in range(1,11):
    url ='https://datascience.stackexchange.com/questions'
    response = requests.get(url, params={'tab':'Votes', 'page':page})
    print(response.url)
    soup = BeautifulSoup(response.text, 'lxml')
    soup

    soup.prettify
    results = soup.find_all('div', attrs={'class': 's-post-summary--content'})

    for product in results:
        url = 'https://datascience.stackexchange.com'+product.find('a').get('href')
        question = product.find('h3').text.strip()
        response = requests.get(url)    
        soup = BeautifulSoup(response.text, 'html')
        soup

        soup.prettify
        try:
            full_question = soup.find('div', {'class' : 's-prose js-post-body', 'itemprop' : 'text'}).find('p').text
        except AttributeError:
            full_question = None
        questions.append(question)
        full_questions.append(full_question)
        time.sleep(3)
    print(len(questions))
        
df = pd.DataFrame(data={'questions':questions,'full_questions':full_questions})
```
