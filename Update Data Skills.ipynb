```python
import pandas as pd
from sqlalchemy import create_engine
from secrets import settings
```

```python
df_ds = pd.read_csv('skills/dataskill.csv')
df_ds = df_ds.set_index('Skill')
df_ds.head()
```

```python
engine = create_engine(settings['skills_db2'])
```

```python
def update_db(db):
    df = pd.read_sql_query(f'select * from "{db}"', engine)
    for i, row in df.iterrows():
        data_skills = []
        skills = row['skills']
        if skills is None:
            continue
        for s in skills.split('; '):
            if s in df_ds.index:
                data_skills.append(df_ds.loc[s].item())  
        if row['data_skills'] is None and len(data_skills) < 1:
            continue
        if row['data_skills'] is None:
            df.loc[i, 'all_skills'] = '; '.join(data_skills)
        elif len(data_skills) > 1:
            df.loc[i, 'all_skills'] = row['data_skills'] + '; ' + '; '.join(data_skills)
        else:
            df.loc[i, 'all_skills'] = row['data_skills']
    return df
```

```python
def update_db(db):
    df = pd.read_sql_query(f'select * from "{db}"', engine)
    df = df.drop(columns={'all_skills'})
    for i, row in df.iterrows():
        data_skills = []
        skills = row['skills']
        if skills is None:
            continue
        for s in skills.split('; '):
            if s in df_ds.index:
                data_skills.append(df_ds.loc[s].item())  
        if row['data_skills'] is None and len(data_skills) < 1:
            continue
        if row['data_skills'] is None:
            df.loc[i, 'data_skills'] = '; '.join(data_skills)
        elif len(data_skills) > 1:
            df.loc[i, 'data_skills'] = row['data_skills'] + '; ' + '; '.join(data_skills)
        else:
            df.loc[i, 'data_skills'] = row['data_skills']
    return df
```

```python
db = 'ContentJMLR'
df = update_db(db)
print(len(df))
df.to_sql(db, engine, index=False, if_exists='replace')
df.head()
```
