```python
import pandas as pd
from skill_extraction import extract_skills, extract_ignore
```

```python
df = pd.read_csv('results/contents.csv')
df = df[['id', 'title', 'url', 'published_date', 'updated_date', 'content', 'author', 'page_text']]
df.head()
```

```python
for i, row in df.iterrows():
    content = row['page_text']
    if not isinstance(content, str):
        continue
    all_skills = extract_skills(content)
    keep_skills, _ = extract_ignore(all_skills)
    keep_skills.sort()
    df.loc[i, 'skills'] = '; '.join(keep_skills)

df.head()
```

```python
for i, row in df.loc[df['skills'].isna()].iterrows():
    content = row['page_text']
    if not isinstance(content, str):
        content = row['content']
    if not isinstance(content, str):
        continue
    all_skills = extract_skills(content)
    keep_skills, _ = extract_ignore(all_skills)
    keep_skills.sort()
    df.loc[i, 'skills'] = '; '.join(keep_skills)

df.head()
```

```python
df.loc[df['skills'].isna()]
```

```python
def safe_float_convert(x):
    try:
        float(x)
        return True
    except ValueError:
        return False
    except TypeError:
        return False
```

```python
mask = df['id'].map(safe_float_convert)
df_new = df.loc[mask].copy()
df_new = df_new.drop(columns=['page_text'])
df_new.head()
```

```python
df_new.to_csv('results/contents_skills.csv', index=False)
```
