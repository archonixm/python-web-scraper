## Import packages

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

## URL of the website to scrape

```python
url = 'https://malaysia.indeed.com/jobs?q=Data+Scientist&l='
url
```

```python
page = requests.get(url)
page
```

```python
soup = BeautifulSoup(page.content, 'html.parser')
soup
```

## Get useful information

Go to the website and use Inspect Element to get the class of the items you are trying to scrape.

<img src="images/inspect-element.png" style="height: 480px;"> 

```python
# div class="tapItem"
# This is to get the first 'tapItem' class item
results = soup.find(class_='tapItem')
print(results.prettify())
```

```python
# Get ALL 'tapItem' class items
job_elems = soup.find_all('a', class_='tapItem')

for job_elem in job_elems:
    print(job_elem, end='\n'*2)
```

```python
test_job = job_elems[0]
test_job
```

### Title

Similarly, use Inspect Element to find the HTML Tag (a/h1/h2/etc) and class of the information you are trying to find.

### Description

As we can see on the main page, we cannot get the whole job description from that page itself. We need to go to each to the webpage of each individual job to get the whole description.

```python
# Get URL link to the individual job page
job_url = "https://malaysia.indeed.com" + test_job['href']
page = requests.get(job_url)
job_soup = BeautifulSoup(page.content, 'html.parser')
print(job_soup.prettify())
```

```python
job_soup.find(id="jobDescriptionText").get_text()
```
